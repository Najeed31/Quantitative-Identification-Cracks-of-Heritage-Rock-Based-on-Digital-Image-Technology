```python
import os
from PIL import Image
from IPython.display import display, HTML

def display_images_in_folder(folder_path):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The folder '{folder_path}' does not exist.")
        return

    # Get a list of files in the folder
    files = os.listdir(folder_path)

    # Filter out only the JPG files
    jpg_files = [file for file in files if file.endswith('.jpg')]

    # Check if there are any JPG files
    if not jpg_files:
        print(f"No JPG images found in the '{folder_path}' folder.")
        return

    # Display the JPG images with labels
    for i, jpg_file in enumerate(jpg_files, start=1):
        image_path = os.path.join(folder_path, jpg_file)
        try:
            img = Image.open(image_path)
            display(HTML(f"<h4>Image {i}</h4>"))
            display(img)
        except Exception as e:
            print(f"Error opening image {jpg_file}: {e}")

# Specify the folder path containing the images
folder_path = '/content/drive/MyDrive/data'

# Call the function to display images in the folder
display_images_in_folder(folder_path)
```

#Data Preprocessing - Image Grayscaling

##Component Method

```python
from PIL import Image
import matplotlib.pyplot as plt

def grayscale_component_method(image):
    # Convert image to RGB mode if not already in RGB
    image = image.convert('RGB')

    # Get pixel data
    pixels = image.load()

    # Iterate over each pixel and calculate grayscale value
    width, height = image.size
    for y in range(height):
        for x in range(width):
            # Get RGB values of the pixel
            r, g, b = pixels[x, y]
            # Calculate grayscale value using component method
            gray_value = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            # Set pixel to grayscale value
            pixels[x, y] = (gray_value, gray_value, gray_value)

    return image

# Load an image
input_image_path = '/content/drive/MyDrive/data/00027.jpg'

# Open image file
input_image = Image.open(input_image_path)

# Perform grayscale conversion using component method
output_image = grayscale_component_method(input_image)

# Display the output image
plt.imshow(output_image)
plt.axis('off')
plt.show()
```

##Maximum Method

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def grayscale_maximum_method(image):
    # Convert image to RGB mode if not already in RGB
    image = image.convert('RGB')

    # Convert image to numpy array
    image_array = np.array(image)

    # Extract red, green, and blue channels
    red_channel = image_array[:, :, 0]
    green_channel = image_array[:, :, 1]
    blue_channel = image_array[:, :, 2]

    # Calculate grayscale using maximum method
    grayscale_image = np.maximum(red_channel, green_channel, blue_channel)

    # Convert grayscale image array back to image
    grayscale_image = Image.fromarray(grayscale_image.astype('uint8'))

    return grayscale_image

# Load an image
input_image_path = '/content/drive/MyDrive/data/00027.jpg'

# Open image file
input_image = Image.open(input_image_path)

# Perform grayscale conversion using maximum method
output_image = grayscale_maximum_method(input_image)

# Display the output image
plt.imshow(output_image, cmap='gray')
plt.axis('off')
plt.show()
```

##Average Method

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def grayscale_average_method(image):
    # Convert image to RGB mode if not already in RGB
    image = image.convert('RGB')

    # Convert image to numpy array
    image_array = np.array(image)

    # Extract red, green, and blue channels
    red_channel = image_array[:, :, 0]
    green_channel = image_array[:, :, 1]
    blue_channel = image_array[:, :, 2]

    # Calculate grayscale using average method
    grayscale_image = (red_channel + green_channel + blue_channel) // 3

    # Convert grayscale image array back to image
    grayscale_image = Image.fromarray(grayscale_image.astype('uint8'))

    return grayscale_image

# Load an image
input_image_path = '/content/drive/MyDrive/data/00027.jpg'

# Open image file
input_image = Image.open(input_image_path)

# Perform grayscale conversion using average method
output_image = grayscale_average_method(input_image)

# Display the output image
plt.imshow(output_image, cmap='gray')
plt.axis('off')
plt.show()
```

##Weighted Average Method

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def grayscale_weighted_average_method(image):
    # Convert image to RGB mode if not already in RGB
    image = image.convert('RGB')

    # Convert image to numpy array
    image_array = np.array(image)

    # Extract red, green, and blue channels
    red_channel = image_array[:, :, 0]
    green_channel = image_array[:, :, 1]
    blue_channel = image_array[:, :, 2]

    # Calculate grayscale using weighted average method
    grayscale_image = 0.3 * red_channel + 0.59 * green_channel + 0.11 * blue_channel

    # Convert grayscale image array back to image
    grayscale_image = Image.fromarray(grayscale_image.astype('uint8'))

    return grayscale_image

# Load an image
# input_image_path = '/content/drive/MyDrive/data/00017.jpg'
# input_image_path = '/content/drive/MyDrive/data/00021.jpg'
# input_image_path = '/content/drive/MyDrive/data/00027.jpg'
# input_image_path = '/content/drive/MyDrive/data/00031.jpg'
input_image_path = '/content/drive/MyDrive/data/00330.jpg'
# Open image file
input_image = Image.open(input_image_path)

# Perform grayscale conversion using weighted average method
output_image = grayscale_weighted_average_method(input_image)

# Save the output image
# output_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00017.jpg'
# output_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00021.jpg'
# output_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00027.jpg'
# output_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00031.jpg'
output_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00330.jpg'
output_image.save(output_image_path)

# Display the output image
plt.imshow(output_image, cmap='gray')
plt.axis('off')
plt.show()

print(f"Grayscale image saved as {output_image_path}.")
```

#Image Filtering

##Mean Filtering

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def mean_filter(image, kernel_size):
    # Convert image to numpy array
    image_array = np.array(image)

    # Get image dimensions
    height, width = image_array.shape

    # Create an empty array to store the filtered image
    filtered_image = np.zeros_like(image_array)

    # Calculate padding size
    pad_size = kernel_size // 2

    # Pad the image with zeros
    padded_image = np.pad(image_array, ((pad_size, pad_size), (pad_size, pad_size)), mode='constant')

    # Apply mean filtering
    for i in range(pad_size, height + pad_size):
        for j in range(pad_size, width + pad_size):
            # Extract the kernel
            kernel = padded_image[i - pad_size:i + pad_size + 1, j - pad_size:j + pad_size + 1]
            # Calculate the mean value of the kernel
            filtered_image[i - pad_size, j - pad_size] = np.mean(kernel)

    return filtered_image

# Load grayscale image
input_image_path = '/content/drive/MyDrive/data/grayscaled_images/gs_00017.jpg'
input_image = Image.open(input_image_path).convert('L')  # Convert to grayscale

# Apply mean filtering with a kernel size of 3x3
kernel_size = 3
filtered_image = mean_filter(input_image, kernel_size)

# Convert filtered image array back to image
output_image = Image.fromarray(filtered_image.astype('uint8'))

# Display the original and filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(output_image, cmap='gray')
plt.title('Mean Filtered Image')
plt.axis('off')

plt.show()
```

#Fracture Segmentation

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def region_growing(image, seed, threshold):
    # Convert image to numpy array
    image_array = np.array(image)

    # Create a mask to keep track of segmented pixels
    segmented = np.zeros_like(image_array)

    # Define neighbors indices (4-connectivity)
    neighbors_indices = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    # Define seed coordinates
    seed_x, seed_y = seed

    # Define seed value
    seed_value = image_array[seed_x, seed_y]

    # Initialize queue with seed pixel
    queue = [(seed_x, seed_y)]

    # Region growing loop
    while queue:
        # Get current pixel coordinates
        current_x, current_y = queue.pop(0)

        # Check if current pixel is within image bounds and not yet segmented
        if (0 <= current_x < image_array.shape[0] and
            0 <= current_y < image_array.shape[1] and
            segmented[current_x, current_y] == 0):

            # Check if current pixel gray value is within threshold range
            if abs(image_array[current_x, current_y] - seed_value) <= threshold:
                # Add current pixel to segmented region
                segmented[current_x, current_y] = 255

                # Add neighbors to queue
                for dx, dy in neighbors_indices:
                    queue.append((current_x + dx, current_y + dy))

    # Convert segmented array back to image
    segmented_image = Image.fromarray(segmented.astype('uint8'))

    return segmented_image

# Load the original image
# original_image_path = '/content/drive/MyDrive/data/binarized_images/binarized_00017.jpg'
# original_image_path = '/content/drive/MyDrive/data/binarized_images/binarized_00021.jpg'
# original_image_path = '/content/drive/MyDrive/data/binarized_images/binarized_00027.jpg'
original_image_path = '/content/drive/MyDrive/data/binarized_images/binarized_00031.jpg'
# original_image_path = '/content/drive/MyDrive/data/binarized_images/binarized_00330.jpg'

original_image = Image.open(original_image_path).convert('L')  # Convert to grayscale if not already

# Define seed coordinates (you may need to adjust this)
seed_coordinates = (100, 0)

# Define threshold value (you may need to adjust this)
threshold_value = 20

# Perform region growing segmentation
segmented_image = region_growing(original_image, seed_coordinates, threshold_value)


output_image =segmented_image
# output_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00017.jpg'
# output_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00021.jpg'
# output_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00027.jpg'
output_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00031.jpg'
# output_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00330.jpg'
output_image.save(output_image_path)

# Display the original and segmented images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')

plt.show()
```

#Morphological Processing

```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def morphological_corrosion(image, kernel_size):
    # Convert image to numpy array
    image_array = np.array(image)

    # Define the structuring element (kernel) for erosion
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    # Perform erosion operation
    eroded_image_array = cv2.erode(image_array, kernel, iterations=1)

    # Convert eroded image array back to image
    eroded_image = Image.fromarray(eroded_image_array.astype('uint8'))

    return eroded_image

# Load the segmented image (from the previous step)
# segmented_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00017.jpg'
# segmented_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00021.jpg'
# segmented_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00027.jpg'
segmented_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00031.jpg'
# segmented_image_path = '/content/drive/MyDrive/data/segmented_images/segmented_00330.jpg'

segmented_image = Image.open(segmented_image_path).convert('L')  # Convert to grayscale if not already

# Define kernel size for erosion (you may need to adjust this)
kernel_size = 3

# Perform morphological corrosion
corroded_image = morphological_corrosion(segmented_image, kernel_size)


output_image =segmented_image
# output_image_path = '/content/drive/MyDrive/data/crack_extracted/crack_00017.jpg'
# output_image_path = '/content/drive/MyDrive/data/crack_extracted/crack_00021.jpg'
# output_image_path = '/content/drive/MyDrive/data/crack_extracted/crack_00027.jpg'
output_image_path = '/content/drive/MyDrive/data/crack_extracted/crack_00031.jpg'
# output_image_path = '/content/drive/MyDrive/data/crack_extracted/crack_00330.jpg'

output_image.save(output_image_path)

# Display the segmented and corroded images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(corroded_image, cmap='gray')
plt.title('Corroded Image')
plt.axis('off')

plt.show()
```
